```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# Your existing preprocessing code remains the same until model training
categorical_features = ['Type', 'Age_group']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(data_new[categorical_features])
encoded_features = encoder.transform(data_new[categorical_features])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))
numerical_features = data_new.drop(categorical_features + ['Total'], axis='columns')
x = pd.concat([numerical_features, encoded_df], axis='columns')
y = data_new.Total

# Split and train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
lr = LinearRegression()
lr.fit(x_train, y_train)

# Make predictions
y_pred = lr.predict(x_test)

# 2. Classification Metrics (after converting to binary classification)
def print_classification_metrics(y_true, y_pred):
    # Convert continuous values to binary classes based on median
    median = np.median(y_true)
    y_true_binary = (y_true > median).astype(int)
    y_pred_binary = (y_pred > median).astype(int)

    accuracy = accuracy_score(y_true_binary, y_pred_binary)
    precision = precision_score(y_true_binary, y_pred_binary)
    recall = recall_score(y_true_binary, y_pred_binary)
    f1 = f1_score(y_true_binary, y_pred_binary)

    print("\nClassification Metrics (after binary conversion):")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

print_classification_metrics(y_test, y_pred)

# If you want to get the scores as variables instead of printing them:
def get_all_metrics(y_true, y_pred):
    # Regression metrics
    regression_metrics = {
        'mae': mean_absolute_error(y_true, y_pred),
        'mse': mean_squared_error(y_true, y_pred),
        'rmse': np.sqrt(mean_squared_error(y_true, y_pred)),
        'r2': r2_score(y_true, y_pred)
    }

    # Classification metrics (binary conversion)
    median = np.median(y_true)
    y_true_binary = (y_true > median).astype(int)
    y_pred_binary = (y_pred > median).astype(int)

    classification_metrics = {
        'accuracy': accuracy_score(y_true_binary, y_pred_binary),
        'precision': precision_score(y_true_binary, y_pred_binary),
        'recall': recall_score(y_true_binary, y_pred_binary),
        'f1': f1_score(y_true_binary, y_pred_binary)
    }

    return regression_metrics, classification_metrics

# Example usage:
regression_metrics, classification_metrics = get_all_metrics(y_test, y_pred)
```

##Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, r2_score

# Preprocessing
categorical_features = ['Type', 'Age_group']
encoder = OneHotEncoder(sparse_output=False)
encoded_features = encoder.fit_transform(data_new[categorical_features])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))

# Combine features
X = pd.concat([data_new.drop(categorical_features + ['Total'], axis=1), encoded_df], axis=1)
y = data_new.Total
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest with optimized parameters
rf_model = RandomForestRegressor(
    n_estimators=10,
    max_depth=None,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Train and time the model
from time import time
start_time = time()
rf_model.fit(X_train, y_train)
training_time = time() - start_time

print(f"\nTraining time: {training_time:.2f} seconds")

# Make predictions
y_pred = rf_model.predict(X_test)

# Convert to binary classification using median as threshold
median = np.median(y_test)
y_test_binary = (y_test > median).astype(int)
y_pred_binary = (y_pred > median).astype(int)

# Calculate metrics
metrics = {
    'R2 Score': r2_score(y_test, y_pred),
    'Accuracy': accuracy_score(y_test_binary, y_pred_binary),
    'Precision': precision_score(y_test_binary, y_pred_binary),
    'Recall': recall_score(y_test_binary, y_pred_binary),
    'F1 Score': f1_score(y_test_binary, y_pred_binary)
}

# Print results
print("\nModel Performance Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```

observation:- using Random foreset classifier we got Accuracy as 0.74 with r2 score as 0.82

##Decision tree

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, r2_score

# Preprocessing
categorical_features = ['Type', 'Age_group']
encoder = OneHotEncoder(sparse_output=False)
encoded_features = encoder.fit_transform(data_new[categorical_features])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))

# Combine features
X = pd.concat([data_new.drop(categorical_features + ['Total'], axis=1), encoded_df], axis=1)
y = data_new.Total
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Decision Tree with basic parameters
dt_model = DecisionTreeRegressor(
    max_depth=10,          # Limit tree depth to prevent overfitting
    min_samples_split=5,   # Minimum samples required to split an internal node
    min_samples_leaf=2,    # Minimum samples required to be at a leaf node
    random_state=42
)

# Train and time the model
from time import time
start_time = time()
dt_model.fit(X_train, y_train)
training_time = time() - start_time

print(f"\nTraining time: {training_time:.2f} seconds")

# Make predictions
y_pred = dt_model.predict(X_test)

# Convert to binary classification using median as threshold
median = np.median(y_test)
y_test_binary = (y_test > median).astype(int)
y_pred_binary = (y_pred > median).astype(int)

# Calculate metrics
metrics = {
    'R2 Score': r2_score(y_test, y_pred),
    'Accuracy': accuracy_score(y_test_binary, y_pred_binary),
    'Precision': precision_score(y_test_binary, y_pred_binary),
    'Recall': recall_score(y_test_binary, y_pred_binary),
    'F1 Score': f1_score(y_test_binary, y_pred_binary)
}

# Print results
print("\nModel Performance Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```
